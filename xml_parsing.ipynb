## Loading files and parse

```python
%%time
path = '/mnt/Разбор XML по УСН/' #11362 files
file_list = glob.glob(path+'*.xml') # if import glob, you can use wildcard: ? *
print('Files found:',len(file_list))

xml_columns = ['ДатаСост', 'ДатаДок', 'ИдДок','НаимОрг','ИННЮЛ','ПризнЕСХН','ПризнУСН','ПризнЕНВД','ПризнСРП']
df_all = pd.DataFrame(columns=xml_columns)

arr_texts = []
for idx_file, file in enumerate(file_list[:]):
    print(idx_file, end='-')
    etree = et.parse(open(file, encoding='utf-8')) #create an ElementTree object 
    
    docs = etree.findall('.//Документ')
        
    for doc in docs:
        arr_text = []
        #print(doc.attrib, doc.tag, doc.getchildren())
        
        arr_text.append(doc.attrib['ДатаСост'])
        arr_text.append(doc.attrib['ДатаДок'])
        arr_text.append(doc.attrib['ИдДок'])
        
        doc_ch = doc.getchildren()
        
        arr_text.append(doc_ch[0].attrib['НаимОрг'])
        arr_text.append(doc_ch[0].attrib['ИННЮЛ'])
        
        arr_text.append(doc_ch[1].attrib['ПризнЕСХН'])
        arr_text.append(doc_ch[1].attrib['ПризнУСН'])
        arr_text.append(doc_ch[1].attrib['ПризнЕНВД'])
        arr_text.append(doc_ch[1].attrib['ПризнСРП'])
        
        arr_texts.append(arr_text)
        
    print(len(arr_texts), end=' ')
        
df_all = pd.DataFrame(data=arr_texts, columns=xml_columns)
#df_all = pd.concat([df_all, df])
    
df_all = df_all.reset_index(drop=True)
print('\n', df_all.shape)

#WRITE

#df_all[:990000].to_excel(path+'all_part1_'+str(len(df_all))+'.xlsx')
#df_all[990000:].to_excel(path+'all_part2_'+str(len(df_all))+'.xlsx')
```
